```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, BatchNormalization, Activation, LSTM, Bidirectional, Dense, Dropout
from keras.metrics import AUC
from sklearn.metrics import accuracy_score, matthews_corrcoef, cohen_kappa_score, roc_auc_score, confusion_matrix, average_precision_score

# Load your data (replace with actual data loading code)
tdf1 = dfM  # Replace with actual train positive dataset
df1 = tdfM  # Replace with actual train negative dataset

# Combine train and test data for cross-validation
X = pd.concat([df1.drop('Target', axis=1), tdf1.drop('Target', axis=1)], axis=0)
y = pd.concat([df1['Target'], tdf1['Target']], axis=0)

# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Reshape for CNN input
X = X.reshape(X.shape[0], X.shape[1], 1)

# Initialize K-fold cross-validation
n_splits = 10  # You can adjust the number of splits
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Create full CNN-BiLSTM model
def create_full_model(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=128, kernel_size=3))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=64, kernel_size=3))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Bidirectional(LSTM(128, return_sequences=True)))
    model.add(Bidirectional(LSTM(64)))
    model.add(Dense(64))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', AUC()])
    return model

# Create model without BatchNormalization
def create_model_no_bn(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, input_shape=input_shape))
    model.add(Activation('sigmoid'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=128, kernel_size=3))
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=64, kernel_size=3))
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Bidirectional(LSTM(128, return_sequences=True)))
    model.add(Bidirectional(LSTM(64)))
    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', AUC()])
    return model

# Create model without Dropout
def create_model_no_dropout(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=128, kernel_size=3))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=64, kernel_size=3))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Bidirectional(LSTM(128, return_sequences=True)))
    model.add(Bidirectional(LSTM(64)))
    model.add(Dense(64))
    model.add(BatchNormalization())
    model.add(Activation('relu'))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', AUC()])
    return model

# Create smaller model
def create_small_model(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=32, kernel_size=3, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=64, kernel_size=3))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Bidirectional(LSTM(64, return_sequences=True)))
    model.add(Bidirectional(LSTM(32)))
    model.add(Dense(32))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', AUC()])
    return model

# Store the models in a dictionary
models = {
    'Full Model': create_full_model(X.shape[1:]),
    'No BatchNorm': create_model_no_bn(X.shape[1:]),
    'No Dropout': create_model_no_dropout(X.shape[1:]),
    'Small Model': create_small_model(X.shape[1:])
}

# Results dictionary
results = {}

# K-fold Cross-validation for each model
for model_name, model_variant in models.items():
    print(f"Training {model_name}")

    # Lists to store evaluation metrics
    accuracies = []
    mccs = []
    kappas = []
    specificities = []
    sensitivities = []
    aurocs = []
    average_precisions = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Train the model
        model_variant.fit(X_train, y_train, epochs=100, batch_size=128, verbose=0)

        # Generate predictions on the test data
        y_pred = model_variant.predict(X_test)
        y_pred_binary = (y_pred > 0.5).astype(int)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred_binary)
        mcc = matthews_corrcoef(y_test, y_pred_binary)
        kappa = cohen_kappa_score(y_test, y_pred_binary)
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred_binary).ravel()
        specificity = tn / (tn + fp)
        sensitivity = tp / (tp + fn)
        auroc = roc_auc_score(y_test, y_pred)
        average_precision = average_precision_score(y_test, y_pred)

        # Append metrics to lists
        accuracies.append(accuracy)
        mccs.append(mcc)
        kappas.append(kappa)
        specificities.append(specificity)
        sensitivities.append(sensitivity)
        aurocs.append(auroc)
        average_precisions.append(average_precision)

    # Save results for comparison
    results[model_name] = {
        'Average Accuracy': np.mean(accuracies),
        'Average MCC': np.mean(mccs),
        'Average Kappa': np.mean(kappas),
        'Average Specificity': np.mean(specificities),
        'Average Sensitivity': np.mean(sensitivities),
        'Average AUROC': np.mean(aurocs),
        'Average Average Precision': np.mean(average_precisions)
    }

# Print results of the ablation study
for model_name, metrics in results.items():
    print(f"\n{model_name} Results:")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
```

```python
from Bio.SeqUtils import ProtParamData

def compute_hydrophobicity(sequences, label):
    if not sequences:
        print(f"Error: No sequences found for {label}!")
        return

    total_hydrophobicity = 0

    for seq in sequences:
        valid_aa = [aa for aa in seq if aa in ProtParamData.kd]  # Filter valid amino acids
        if not valid_aa:
            continue
        total_hydrophobicity += sum(ProtParamData.kd[aa] for aa in valid_aa) / len(valid_aa)

    avg_hydrophobicity = total_hydrophobicity / len(sequences)
    print(f"{label} Sequences - Average Hydrophobicity: {avg_hydrophobicity:.2f}")

# Compute hydrophobicity
compute_hydrophobicity(pos_sequences, "Positive")
compute_hydrophobicity(neg_sequences, "Negative")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential, load_model
from keras.layers import Conv1D, MaxPooling1D, Flatten, BatchNormalization, Activation, LSTM, Bidirectional, Dense, Dropout
from keras.metrics import AUC
from sklearn.metrics import accuracy_score, matthews_corrcoef, cohen_kappa_score, roc_auc_score, confusion_matrix, average_precision_score

# Load the training data
df1 = dfM
tdf1 = tdfM

# Combine train and test data for training
X_train = pd.concat([df1.drop('Target', axis=1), tdf1.drop('Target', axis=1)], axis=0)
y_train = pd.concat([df1['Target'], tdf1['Target']], axis=0)

# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# Reshape data for CNN-BiLSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

# Create a CNN-BiLSTM model
model = Sequential()

model.add(Conv1D(filters=64, kernel_size=3, input_shape=(X_train.shape[1], 1)))
model.add(BatchNormalization())
model.add(Activation('sigmoid'))
model.add(MaxPooling1D(pool_size=2))

model.add(Conv1D(filters=128, kernel_size=3))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling1D(pool_size=2))

model.add(Conv1D(filters=64, kernel_size=3))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling1D(pool_size=2))

model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(64))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', AUC()])

# Train the model
model.fit(X_train, y_train, epochs=250, batch_size=128, verbose=1)

# Save the trained model
model.save("cnn_bilstm_model.h5")

# ---------------- TESTING WITH AD.CSV ----------------

# Load the test data
AD = pd.read_csv('/home/shaon/Downloads/AD1.csv')

# Preprocess test data
X_test = AD.drop('Target', axis=1)
y_test = AD['Target']

X_test = scaler.transform(X_test)  # Use the same scaler from training
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)  # Reshape for CNN-BiLSTM

# Load the trained model
model = load_model("cnn_bilstm_model.h5")

# Generate predictions
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred_binary)
mcc = matthews_corrcoef(y_test, y_pred_binary)
kappa = cohen_kappa_score(y_test, y_pred_binary)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_binary).ravel()
specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)
auroc = roc_auc_score(y_test, y_pred)
average_precision = average_precision_score(y_test, y_pred)

# Print results
print(f'Accuracy: {accuracy:.4f}')
print(f'MCC: {mcc:.4f}')
print(f'Kappa Score: {kappa:.4f}')
print(f'Specificity: {specificity:.4f}')
print(f'Sensitivity: {sensitivity:.4f}')
print(f'AUROC: {auroc:.4f}')
print(f'Average Precision: {average_precision:.4f}')
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# Load the dataset
AD = pd.read_csv('/home/shaon/Downloads/AD1.csv')

# Split features and target
X = AD.drop(columns=['Target'])  # Features
y = AD['Target']  # Target column

# Apply SMOTE to balance the dataset
smote = SMOTE(sampling_strategy=1.0, random_state=42)  # 1.0 means equal balancing
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame with oversampled data
AD_balanced = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Target'])], axis=1)

# Check class distribution after oversampling
print(AD_balanced['Target'].value_counts())  # Should print 64 for both classes

# Save the balanced dataset
AD_balanced.to_csv('/home/shaon/Downloads/AD_balanced.csv', index=False)

print("\nOversampling complete! Balanced dataset saved as 'AD_balanced.csv'.")
```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# Load dataset
AD = pd.read_csv('/home/shaon/Downloads/AD1.csv')

# Split features and target
X = AD.drop(columns=['Target'])  # Features
y = AD['Target']  # Target column

# Apply undersampling: Reduce class 0 (majority) to 30 while keeping all class 1 samples
undersampler = RandomUnderSampler(sampling_strategy={0: 30, 1: 30}, random_state=42)
X_resampled, y_resampled = undersampler.fit_resample(X, y)

# Create a new DataFrame with undersampled data
AD_undersampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Target'])], axis=1)

# Check class distribution after undersampling
print(AD_undersampled['Target'].value_counts())  # Should print 30 for both classes

# Save the undersampled dataset
AD_undersampled.to_csv('/home/shaon/Downloads/AD_undersampled.csv', index=False)

print("\nUndersampling complete! Balanced dataset saved as 'AD_undersampled.csv'.")
```
